```python
# ==============================================
# IA DE YOUTUBERS
# Uso de:
# - Lista de diccionarios
# - if
# - elif
# - else
# - while True
# - for
# ==============================================

print("=========================================")
print("      IA DE LOS YOUTUBERS VIRTUALES")
print("       CBTIS 128 - Programacion")
print("=========================================")

# Lista vacia
los_youtubers = []

# Diccionario vacio
youtuber = {}

while True:

    print("\n=========== MENU ==========")
    print("1. Registrar youtuber")
    print("2. Ver los youtubers")
    print("3. Analizar canal")
    print("4. Eliminar ultimo youtuber")
    print("5. Salir")
    print("===========================")

    opcion = int(input("Selecciona una opcion: "))

    # OPCION 1
    if opcion == 1:

        nombre = input("Nombre del youtuber: ")
        categoria = input("Categoria del canal: ")
        seguidores = int(input("Numero de seguidores: "))

        # Diccionario
        youtuber = {
            "nombre": nombre,
            "categoria": categoria,
            "seguidores": seguidores
        }

        # Guardar en la lista
        los_youtubers.append(youtuber)

        print("Youtuber registrado correctamente")

    # OPCION 2
    elif opcion == 2:

        if los_youtubers == []:
            print("No hay los youtubers registrados")

        else:

            print("\n+------------------------------------------------------+")
            print("| ID | NOMBRE | CATEGORIA | SEGUIDORES               |")
            print("+------------------------------------------------------+")

            contador = 1

            for datos in los_youtubers:

                print("|", contador,
                      "|", datos["nombre"],
                      "|", datos["categoria"],
                      "|", datos["seguidores"])

                contador = contador + 1

            print("+------------------------------------------------------+")

    # OPCION 3
    elif opcion == 3:

        cantidad = len(los_youtubers)

        print("\n=========== ANALISIS IA ==========")

        if cantidad == 0:
            print("No existen canales registrados")
            print("La IA recomienda agregar los youtubers")

        elif cantidad >= 1 and cantidad <= 3:
            print("Pocos los youtubers registrados")
            print("La IA detecta canales principiantes")

        elif cantidad >= 4 and cantidad <= 6:
            print("Comunidad en crecimiento")
            print("La IA detecta potencial viral")

        else:
            print("Muchos los youtubers registrados")
            print("La IA detecta futuros influencers")

    # OPCION 4
    elif opcion == 4:

        if los_youtubers == []:
            print("No hay los youtubers para eliminar")

        else:

            eliminado = los_youtubers.pop()

            print("Youtuber eliminado:", eliminado["nombre"])

    # OPCION 5
    elif opcion == 5:

        print("Cerrando sistema de IA...")
        print("Hasta luego")
        break

    # OPCION INVALIDA
    else:

        print("Opcion incorrecta")
```
